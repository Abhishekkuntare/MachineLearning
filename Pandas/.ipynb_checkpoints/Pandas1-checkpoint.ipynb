```python
# 5 - Filtering dataframe on a condition | Masking using boolean series-----------------------
```

```python
data['city']=='Hyderabad' # and it type is series
```

```python
mask = data['city']=='Hyderabad'
data[mask] # in this syntax we get those data which has city name is hyderadbad
```

```python
#create and funcition
def get_city_count(city):
    mask= data['city']==city
    return data[mask].shape[0]#shape 0 means print all the rows which contains the city name
```

```python
get_city_count('Pune')
```

```python
def get_season_match(season):
    mask = data['season']==season
    return data[mask].shape[0]
```

```python
ans = get_season_match(2008)
print(ans,"Matches are Played")
```
